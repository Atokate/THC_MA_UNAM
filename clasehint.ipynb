```python
import sympy as sm
import numpy as np
```

```python
x,y = sm.symbols('x y')
```

```python
a=3
b=4
c=1
d=2
e=3
f=1
```

```python
eq1 = sm.Eq(a*x+b*y,e)
```

```python
eq2 = sm.Eq(c*x+d*y,f)
```

```python
despeje1 = sm.solve(eq1,y)
```

```python
despeje2 = sm.solve(eq2,y)
```

```python
f1 = sm.lambdify(x,despeje1)
```

```python
f2 = sm.lambdify(x,despeje2)
```

```python
f1(25)
```

```python
f1(np.array([25,3,24]))
```

```python
xrango = np.arange(-10,10,.5)
len(xrango)
```

```python
import matplotlib.pyplot as plt 
plt.figure()#empieza el plano
plt.plot(xrango,f1(xrango)[0],'magenta',label = 'funcion 1',marker='o',linestyle =  'none') #subgrafica
plt.plot(xrango,f2(xrango)[0],'blue',label = 'funcion 2')
plt.legend()
plt.grid()


plt.show()#termina el plano
```
