```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xpoints = np.arange(0,10,0.01)#-10 a 10
ypoints = np.linspace(-10,10,1000)
len(ypoints)
```

```python
seno = np.sin(xpoints)
seno
np.exp(xpoints)
```

```python
plt.plot(xpoints,np.exp(xpoints),color = 'b') 
plt.plot(xpoints,ypoints,color = 'g')
plt.plot(xpoints,seno,color = 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,10)
plt.ylim(-10,10)
plt.grid()
plt.title('Graficas de sen(x), exp() y recta')
plt.show
```
