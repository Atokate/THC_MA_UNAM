## 5. Genera un array que, dependiendo de un valor n, contenga los n ́umeros primos comprendidos entre 2 y n.

```python
numprimos = []
j = 0
n = int(input('Ingrese un numero entero para delimitar el rango de busqueda de primos: '))

for i in range(2,n+1):
    primo = True
    for j in range(2,i):
        if i%j == 0:
            primo = False
        elif i == j:
            break
        else:
            continue
    if primo == True:
        numprimos.append(i)
        
print(numprimos)
```

## 6. Usando While realiza una funci ́on multiplicaci ́on, que calcule por ejemplo 5*7 pero usando sumas sucesivas, es decir, suma 5 siete veces.

```python
print('Este es un programa que multiplica dos numero')
a = int(input('Ingrese el primer numero a multiplicar: '))
b = int(input('Ingrese el primer numero a multiplicar: '))
conta = 0
multi = 0

while conta < b:
    multi+= a
    conta +=1
    
print(a,'x',b,'=',multi)
```

## 7. Una pel ́ıcula tiene un presupuesto inicial 150 millones de Euros. La primer semana de producci ́on tiene un costo total de 31 millones. Cada semana que pasa la factura es el 20% inferior a la semana anterior. Proporcione un algoritmo que calcule el n ́umero de semanas que se puede permitir la producci ́on de la pel ́ıcula.

```python
presac = 150000000 #presupuesto actual
costosemanal = 31000000
semanas = 0

while presac > 0 :
    presac -= costosemanal
    costosemanal *= 0.8
    semanas+=1

print('La pelicula se puede permitir',semanas,'semanas de produccion')
```
