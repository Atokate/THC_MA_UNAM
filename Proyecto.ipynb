```python
import matplotlib.pyplot as plt
import numpy as np
```

## Grafica de la emision de CO2 de la quema de combustible

```python
kmelec = np.arange(0,100,.1)
gelec = kmelec*85.85

plt.figure()
plt.plot(gelec,kmelec, color = 'red')
plt.xlabel('km recorridos')
plt.ylabel('Gramos de CO2 emitidos')
plt.title('Grafica los gramos de CO2 emitidos en funcion de los km recorridos por un auto electrico')
plt.grid()
plt.show()
```

## Grafica de la emision de  CO2 por autos de petroleo

```python
kmpet = np.arange(0,100,.1)
gpet = kmpet*143

plt.figure()
plt.plot(gpet,kmpet, color = 'blue')
plt.xlabel('km recorridos')
plt.ylabel('Gramos de CO2 emitidos')
plt.title('Grafica los gramos de CO2 emitidos en funcion de los km recorridos por un auto comun')
plt.grid()
plt.show()
```

## Las dos graficas superpuestas

```python

plt.figure()
plt.plot(gpet,kmpet, color = 'blue', label='Auto con combustible de petroleo')
plt.plot(gelec,kmelec, color = 'red', label='Auto electrico')
plt.xlabel('km recorridos')
plt.ylabel('Gramos de CO2 emitidos')
plt.title('Comparasion de ambas graficas')
plt.xlim(0,8000)
plt.grid()
plt.legend()
plt.show()
```
